```python
import xml.etree.ElementTree as ET
import re
import os
import pickle
#根據xml　檔案進行解析
def extract_pubmed_info(xml_file_path):
    try:
        # 解析XML文件
        tree = ET.parse(xml_file_path)
        root = tree.getroot()

        # 提取所需的信息
        PMID = root.find(".//PMID")
        PMID_text = PMID.text.strip() 
        article_title_element = root.find(".//ArticleTitle")

        article_title = article_title_element.text.strip() if article_title_element is not None else None

        date_revised_element = root.find(".//MedlineCitation/DateRevised/Year")
        date_revised_year = date_revised_element.text.strip() if date_revised_element is not None else None

        date_revised_element = root.find(".//MedlineCitation/DateRevised/Month")
        date_revised_month = date_revised_element.text.strip() if date_revised_element is not None else None

        date_revised_element = root.find(".//MedlineCitation/DateRevised/Day")
        date_revised_day = date_revised_element.text.strip() if date_revised_element is not None else None

        date_revised = date_revised_year + '/' + date_revised_month + '/' + date_revised_day

        abstract_text_element = root.find(".//Abstract")
        abstract_text = ET.tostring(abstract_text_element, encoding='utf-8', method='xml').decode('utf-8')
        abstract_text = re.sub(r'<.*?>', '', abstract_text)


        authors = []
        author_elements = root.findall(".//Author")
        for author_element in author_elements:
            last_name_element = author_element.find(".//LastName")
            first_name_element = author_element.find(".//ForeName")
            last_name = last_name_element.text.strip() if last_name_element is not None else ''
            first_name = first_name_element.text.strip() if first_name_element is not None else ''
            authors.append(f"{last_name}, {first_name}")
         

        pubmed_data = {
            'PMID' : PMID_text,
            'ArticleTitle': article_title,
            'DateRevised': date_revised,
            'AbstractText': abstract_text,
            'Authors': authors
        }
        
        another = text_statistics(pubmed_data["AbstractText"])
        pubmed_data = {**pubmed_data, **another}
        xml_pickle_path = 'xml_pickle/'
        with open(xml_pickle_path+str(pubmed_data["PMID"])+".pkl",'wb') as file :
            pickle.dump(pubmed_data,file)

        return pubmed_data

    except Exception as e:
        print(f"提取信息时出现错误：{str(e)}")
        return None

# 替换下面的路径为你的PubMed XML文件的位置
xml_file_path = os.path.join("teacher","10074110.xml")
print(xml_file_path)
# 调用函数提取信息
result = extract_pubmed_info(xml_file_path)

if result:
    for key, value in result.items():
        print(f"{key}: {value}")
else:
    print("未提取任何信息。")
```

```python
import string
nltk.download('punkt')
def text_statistics(text):
    # 1. 計算字元數
    char_count = len(text)

    # 2. 計算不含空白的字元數
    char_count_no_spaces = len(text.replace(" ", ""))

    # 3. 計算詞數
    text_without_punctuation = text.translate(str.maketrans('', '', string.punctuation))
    words = word_tokenize(text_without_punctuation)
    word_count = len(words)

    # 4. 計算句數
    sentences = sent_tokenize(text)
    sentence_count = len(sentences)

    return {
        "char": char_count,
        "char(no_space)": char_count_no_spaces,
        "words": word_count,
        "sen": sentence_count
    }

# 測試函數
```

```python
def find_dicts_containing_string(list_of_dicts, search_string):
    # 用來存儲包含搜索字符串的字典的列表
    matching_dicts = []

    # 遍歷列表中的每個字典，檢查是否包含搜索字符串
    for dictionary in list_of_dicts:
        for key in ["ArticleTitle",'DateRevised','AbstractText','Authors'] :
            if search_string in dictionary.get(key, ""):
                matching_dicts.append(dictionary)
                break
            

    return matching_dicts
```

```python
def find_dicts_containing_string_num(list_of_dicts, search_string):

    return_dict = list(list_of_dicts)
    for dictionary in return_dict:
        count = 0
        for key in ["ArticleTitle",'DateRevised','AbstractText','Authors']  :
            count += dictionary[key].count(search_string)
        dictionary["matching_num"] = count

    return return_dict
```

```python
def sort_by_num(dictionary):
    return dictionary["matching_num"]
```

```python
import os
listdir = os.listdir("xml")

def parsing_all(match_string):
    temp = os.listdir("xml")
    xml_path = ["xml/" + element for element in temp]
    Article_list = []
    for file_name in xml_path:
        article = extract_pubmed_info(file_name)
        Article_list.append(article)

    matching_dicts = find_dicts_containing_string(Article_list,match_string)
    matching_dicts_and_num = find_dicts_containing_string_num(matching_dicts,match_string)
    sorted_list = sorted(matching_dicts_and_num, key=sort_by_num , reverse= True)
    for k in sorted_list :
        print(k)



    


parsing_all("a")
```
